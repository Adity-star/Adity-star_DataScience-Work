```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Datasets/Zomato/Country-Code.xlsx')
df_country.head()
```

```python
df_final = pd.merge(df,df_country,on='Country Code')
```

```python
df_final.head()
```

```python
country_names = df_final.Country.value_counts().index
```

```python
country_values = df_final.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

observation: Maximum orders are coming form India.

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
plt.figure(figsize = (12,5))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
plt.figure(figsize = (12,5))

sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue = 'Rating color',palette=['blue','red','orange','yellow','green','green','red'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green','red'])
```
